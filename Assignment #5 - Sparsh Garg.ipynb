```python
import glob
import numpy 
import matplotlib
import pandas as pd
import matplotlib.pyplot 
import matplotlib.pyplot as plt
import numpy as np
```

```python
def visualize(filename):
    "Plot the histogram"
    data = pd.read_csv(filepath_or_buffer = filename)
    df_mean= data.mean(axis=0)
    df_mean.plot(xlabel='Feature#', ylabel='Values', title= filename)
    df_max= data.max(axis=0)
    df_max.plot()
    df_min= data.min(axis=0)
    df_min.plot()
    plt.legend(['Average', 'Max', 'Min'])
    plt.show()
```

```python
def stat(filename):
    "Print the descriptive statistics"
    data = pd.read_csv(filepath_or_buffer = filename)
    df_max = data.max(axis=0)
    df_min = data.min(axis=0)
    Sum_min = df_min.sum(axis=0)
    if ((df_max[0] == 0) and (df_max[20] == 12)):
        print("I doubt the integrity of this dataset")
    elif Sum_min == 0:
        print("Why is this happening")
    else:
        print("Dataset looks good")
```

```python
dataset = sorted(glob.glob('data*.csv'))
print(dataset)
```

```python
for datasets in dataset[:4]:
    print("\n")
    print(datasets)
    visualize(datasets)
    stat(datasets)
```
